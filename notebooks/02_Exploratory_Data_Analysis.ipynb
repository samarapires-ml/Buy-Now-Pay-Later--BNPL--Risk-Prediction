```python
import pandas as pd

url = "https://raw.githubusercontent.com/samarapires-ml/Buy-Now-Pay-Later--BNPL--Risk-Prediction/main/data/raw/raw_data.csv"
df = pd.read_csv(url)

df.head()
```

```python
#Correlation Heap Map
import matplotlib.pyplot as plt
import seaborn as sns

numeric_df = df.select_dtypes(include=["int64", "float64"])

plt.figure(figsize=(10, 6))
sns.heatmap(numeric_df.corr(), cmap="Blues", annot=False)
plt.title("Correlation Heatmap")
plt.show()
```

```python
#Numeric Feature Distributions
numeric_cols = df.select_dtypes(include=["int64", "float64"]).columns

df[numeric_cols].hist(figsize=(12, 10), bins=30, edgecolor='black')
plt.suptitle("Numeric Feature Distributions", fontsize=16)
plt.show()
```

```python
#Categorical Feature Distributions
categorical_cols = df.select_dtypes(include=["object"]).columns

plt.figure(figsize=(14, 12))

for i, col in enumerate(categorical_cols, 1):
    plt.subplot(3, 3, i)
    sns.countplot(data=df, x=col)
    plt.xticks(rotation=45)
    plt.title(f"{col} Distribution")

plt.tight_layout()
plt.show()
```
